```python
import pandas as pd
import pandas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('world_population.csv')
df
```

```python
print (df.isnull().sum())
```

```python
df.info()
```

```python
plt.figure(figsize=(16, 10))
sns.barplot(x='Area (km²)', y='Capital', data=df)
plt.xlabel('Area (km²)')
plt.ylabel('Capital')
plt.ylim(25, 50)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='Rank', y='Continent', data=df)
plt.xlabel('Rank')
plt.ylabel('Continent')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='2022 Population', y='Country/Territory', data=df)
plt.xlabel('2022 Population')
plt.ylabel('Country/Territory')
plt.ylim(0, 25)
plt.show()
```

```python
df_raw = df.groupby(['2022 Population', '2020 Population', '2015 Population', '2010 Population', '2000 Population'])
df_raw.first()
```
